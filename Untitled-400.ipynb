```python
from matplotlib import pyplot as plt
import numpy as np
 
 
# Creating dataset
marks_CAP776 = np.array([22, 87, 5, 43, 56,
              73, 55])
 
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
plt.xlabel('Marks')
ax.hist(marks_CAP776, bins = [0, 25, 50, 75, 100])
 
# Show plot
plt.show()
```
