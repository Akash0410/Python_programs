```python
d_patient_profiles = pd.DataFrame(patient_profiles['Age'].replace('None',0).astype('int64'))
dd = d_patient_profiles.loc[:,'Age'].sort_values()
dd
```

```python

l1=0
l2=0
l3=0
l4=0


for i in dd:
    if i > 1 and i <= 40:
        l1+=1
    if i > 40 and i <=70:
        l2+=1
    if i>70:
        l3+=1
    if i ==0:
        l4+=1

age_group = [l1,l2,l3,l4]
age_label = ['1 to 40','41 to 70','Older than 70','People whose age is not Specified']
expl=[0.1,0.2,0.3,0]

cols=['r','orange','g','c']
plt.figure(figsize=(10,8))
plt.pie(age_group, labels= age_label, colors=cols,startangle=0,explode=expl,shadow=True,autopct='%1.0f%%')
plt.legend(title="Age group",loc='lower left')
plt.title(label='Age group of people who got Interacted with Health Camps', fontsize = 20)
plt.show()
```

```python
#Average health score comparision between datasets
sum1=0
for i in First_Health_Camp_Attended['Health_Score']:
    sum1=sum1+i
avg_score1 = (sum1/6216)
print(avg_score1)
```

```python
sum2=0
for j in Second_Health_Camp_Attended['Health Score']:
    sum2=sum2+j
avg_score2 = (sum2/7817)
print(avg_score2)
```

```python
avrgs = [avg_score1,avg_score2]
avrgs_label = ['Average Health Score of First Health Camp','Average Health Score of Second Health Camp']
expl=[0,0.04]
cols=['orange','c']
plt.figure(figsize=(10,8))
plt.pie(avrgs, labels= avrgs_label, colors=cols,startangle=0,explode=expl,shadow=True,autopct='%1.0f%%')
plt.legend(title="Health Score of First and Second Health Camp",loc='lower right')
plt.title(label='Comparision of Health Score of First and Second Health Camp', fontsize = 20)
plt.show()
```
