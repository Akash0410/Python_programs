<h1><i>Different ways to normalize data


```python
df['income'].max() 
```

```python
df['income']=df['income']/df['income'].max()   #simple feature scaling
df.head(5)
```

```python
df['income']=(df['income']-df['income'].min())/(df['income'].max()-df['income'].min())   #min-max
df.head(5)
```

```python
df['income']=(df['income']-df['income'].mean())/(df['income'].std())   #z-score
df.head(5)
```

```python
df['custcat'].value_counts()
```
